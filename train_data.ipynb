```python
import jieba.analyse
import pandas as pd

def get_filename(i):
    str1 = 'K:/download/'
    str2 = '滚动.txt'
    str3 = '原创.txt'
    str4 = '人民娱评.txt'
    str5 = '文艺星青年.txt'
    str6 = '文艺开讲.txt'
    if i < 40:
        file = str1 + str(i) + str2
    else:
        if i < 233:
            file = str1 + str(i) + str3
        else:
            if i < 252:
                file = str1 + str(i) + str4
            else:
                if i < 394:
                    file = str1 + str(i) + str5
                else:
                    file = str1 + str(i) + str6
    return file

splitedStr = ''
segments = []
i = 1
#载入停用词
stopwords = jieba.analyse.set_stop_words('K:/download/stoped.txt')
while i < 450:
    file = get_filename(i)    
    with open(file,encoding='utf-8') as f:
        content = f.read() 
#     words  = jieba.analyse.textrank(content, topK=50,withWeight=False,allowPOS=('ns', 'n', 'vn', 'v'))
    words = jieba.cut(content)
    for word in words :
        # 记录全局分词
        if word not in stopwords:
            segments.append({'word':word, 'count':1})
            splitedStr += word + ' '
            print('分析第' + i + '篇文章，完成！/n')
    i= i + 1
    f.close()
    
dfSg = pd.DataFrame(segments)
# 词频统计
dfWord = dfSg.groupby('word')['count'].sum()
#导出csv
dfWord.to_csv('K:/download/部分停用词.csv',encoding='utf-8')
print('完成')

```
